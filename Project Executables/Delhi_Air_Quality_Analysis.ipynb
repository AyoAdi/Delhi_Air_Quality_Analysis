###Is this dataset perfect or something?

```python
#One thing I dont like is that the months are numbered and i want them to be the names. Let's do that

def renameMonths(rows):
  if rows== 1:
    return 'January'
  elif rows == 2:
    return 'February'
  elif rows == 3:
    return 'March'
  elif rows == 4:
    return 'April'
  elif rows == 5:
    return 'May'
  elif rows == 6:
    return 'June'
  elif rows == 7:
    return 'July'
  elif rows == 8:
    return 'August'
  elif rows == 9:
    return 'September'
  elif rows == 10:
    return 'October'
  elif rows == 11:
    return 'November'
  elif rows == 12:
    return 'December'

new_df = df.copy()

new_df['Month'] = new_df['Month'].apply(renameMonths)
```

```python
#One thing I dont like is that the months are numbered and i want them to be the names. Let's do that

def renameDays(rows):
  if rows== 1:
    return 'Monday'
  elif rows == 2:
    return 'Tuesday'
  elif rows == 3:
    return 'Wednesday'
  elif rows == 4:
    return 'Thursday'
  elif rows == 5:
    return 'Friday'
  elif rows == 6:
    return 'Saturday'
  elif rows == 7:
    return 'Sunday'


new_df['Days'] = new_df['Days'].apply(renameDays)
```
